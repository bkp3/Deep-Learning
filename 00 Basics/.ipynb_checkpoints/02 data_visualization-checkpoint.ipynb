```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.arange(0,10)
```

```python
x
```

```python
y = x**2
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'*')
```

```python
plt.plot(x,y,'red')
```

```python
plt.plot(x,y,'r--')
```

```python
plt.plot(x,y)
plt.xlim(0,4)
plt.ylim(0,10)
plt.title("TITLE")
plt.xlabel('X label')
plt.ylabel('Y label')
```
